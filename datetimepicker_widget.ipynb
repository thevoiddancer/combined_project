```python
import tkinter as tk
import datetime as dt
from random import shuffle

MJESECI = ['Siječanj', 'Veljača', 'Ožujak', 'Travanj', 'Svibanj', 'Lip', 'Srp', 'Kol', 'Ruj', 'Lis', 'Stu', 'Pro']
DANI = ['PON', 'UTO', 'SRI', 'CET', 'PET', 'SUB', 'NED']

def get_days_in_month(year, month):
    today = dt.datetime.now().date()
    next_month = today.replace(day=1, year=year, month=month % 12 + 1)
    day = dt.timedelta(days=1)
    last_day = next_month - day
    number_of_days = last_day.day
    return number_of_days

class DateTimePickerWidget:
    # Promjene: 
    # canvas je dodan kao element objekta (self.date_picker_canvas i self.time_picker_canvas)
    # for petljom prolazimo kroz sve widgete na canvasu i brišemo ih
    # na taj način se poslije svake promjene mjesca ili godine resetira stvar
    # dodan .grid_remove da se time picker sakrije ako se promjeni mjesec ili godina
    # tj. time picker je vidljiv samo nakon što se klikne datum

    def __init__(self, root):
        self.root = root
        now = dt.datetime.now()
        self.today = now.date()
        self.year = self.today.year
        self.month = self.today.month

        self.date_picker_canvas = tk.Canvas(self.root)
        self.date_picker_canvas.grid(row=0, column=0)

        self.create_date_picker()
        #self.create_time_picker()
        self.time_picker_canvas = tk.Canvas(self.root)
        # self.time_picker_canvas.grid(row=0, column=1)

    def create_date_picker(self):
        # date_picker_canvas = tk.Canvas(self.root)
        # date_picker_canvas.grid(row=0, column=0)
        for widget in self.date_picker_canvas.winfo_children():
            widget.destroy()

        label_year = tk.Label(self.date_picker_canvas, text=self.year)
        label_year.grid(row=0, column=1, columnspan=5)

        label_month = tk.Label(self.date_picker_canvas, text=MJESECI[self.month - 1])
        label_month.grid(row=1, column=1, columnspan=5)

        year_prev = tk.Button(self.date_picker_canvas, text='<', command=lambda: self.change_year(-1))
        year_prev.grid(row=0, column=0)
        year_next = tk.Button(self.date_picker_canvas, text='>', command=lambda: self.change_year(1))
        year_next.grid(row=0, column=6)

        month_prev = tk.Button(self.date_picker_canvas, text='<', command=lambda: self.change_month(-1))
        month_prev.grid(row=1, column=0)
        month_next = tk.Button(self.date_picker_canvas, text='>', command=lambda: self.change_month(1))
        month_next.grid(row=1, column=6)

        for index, dan in enumerate(DANI):
            dan_label = tk.Label(self.date_picker_canvas, text=dan)
            dan_label.grid(row=2, column=index)
        
        day_1 = self.today.replace(day=1).weekday()

        for i in range(get_days_in_month(self.year, self.month)):
            day_button = tk.Button(self.date_picker_canvas, text=i + 1, command=lambda dan=i: self.open_time_picker(dan))
            day_button.grid(row=3 + (day_1 + i) // 7, column=(day_1 + i) % 7)

    def open_time_picker(self, dan):
        self.dan = dan
        #termini = get_available_time()
        termini = ['8:00', '9:00', '10:00', '11:00']
        shuffle(termini)
        self.create_time_picker(termini)


    def create_time_picker(self, termini):
        # time_picker_canvas = tk.Canvas(self.root)
        # time_picker_canvas.grid(row=0, column=1)
        self.time_picker_canvas.grid(row=0, column=1)

        label_odaberi = tk.Label(self.time_picker_canvas, text="Odaberi vrijeme")
        label_odaberi.grid(row=0, column=0)

        vrijeme_var = tk.StringVar(self.root)
        vrijeme = tk.OptionMenu(self.time_picker_canvas, vrijeme_var, "8:00", "9:00", "10:00")
        vrijeme.grid(row=1, column=0)

        for index, vrijeme in enumerate(termini):
            vrijeme_gumb = tk.Button(self.time_picker_canvas, text=vrijeme, command=lambda v=vrijeme: self.rezerviraj(v))
            vrijeme_gumb.grid(row=2 + index, column=0)         

        rezerviraj = tk.Button(self.time_picker_canvas, text='Rezerviraj')
        rezerviraj.grid(row=100, column=0)

    def rezerviraj(self, v):
        print(self.year, self.month, self.dan, v)

    def change_month(self, x):
        self.time_picker_canvas.grid_remove()
        if self.month == 1 and x == -1:
            self.month = 12
            self.year -= 1
        elif self.month == 12 and x == 1:
            self.month = 1
            self.year += 1
        else:
            self.month += x
        self.today = self.today.replace(month=self.month, year=self.year)

        self.create_date_picker()
        #self.create_time_picker()

    def change_year(self, x):
        self.time_picker_canvas.grid_remove()
        self.year += x
        self.today = self.today.replace(year=self.year)
        self.create_date_picker()
        #self.create_time_picker()

root = tk.Tk()
app = DateTimePickerWidget(root)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from datetime import datetime, timedelta


class DateTimePickerApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Date & Time Picker")

        # Current date
        self.today = datetime.today()
        self.selected_date = self.today

        # Create the frame
        self.frame = tk.Frame(self.root)
        self.frame.pack(padx=10, pady=10)

        # Left side frame (for date picker)
        self.date_frame = tk.Frame(self.frame)
        self.date_frame.grid(row=0, column=0, padx=10)

        # Right side frame (for time picker)
        self.time_frame = tk.Frame(self.frame)
        self.time_frame.grid(row=0, column=1, padx=10)

        # Date Picker
        self.create_date_picker()

        # Time Picker
        self.create_time_picker()

    def create_date_picker(self):
        """Create the date picker with month and year navigation, and day selection."""
        # Date Picker - Title (Year & Month Selector)
        self.date_title = tk.Label(self.date_frame, text="", font=("Arial", 16))
        self.date_title.grid(row=0, column=0, columnspan=7, padx=5, pady=5)

        # Create month navigation buttons
        self.prev_month_button = tk.Button(self.date_frame, text="<", command=lambda: self.change_month(-1))
        self.prev_month_button.grid(row=0, column=0, padx=5, pady=5)

        self.next_month_button = tk.Button(self.date_frame, text=">", command=lambda: self.change_month(1))
        self.next_month_button.grid(row=0, column=6, padx=5, pady=5)

        # Create year navigation buttons
        self.prev_year_button = tk.Button(self.date_frame, text="<<", command=lambda: self.change_year(-1))
        self.prev_year_button.grid(row=1, column=0, padx=5, pady=5)

        self.next_year_button = tk.Button(self.date_frame, text=">>", command=lambda: self.change_year(1))
        self.next_year_button.grid(row=1, column=6, padx=5, pady=5)

        # Calendar header (day names)
        day_names = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
        for col, day in enumerate(day_names):
            day_label = tk.Label(self.date_frame, text=day, font=("Arial", 10))
            day_label.grid(row=2, column=col, padx=5, pady=5)

        # Create the calendar grid (buttons for each day of the month)
        self.day_buttons = {}
        self.create_calendar()

    def create_time_picker(self):
        """Create the time picker with hourly increments."""
        time_label = tk.Label(self.time_frame, text="Select Time:", font=("Arial", 12))
        time_label.grid(row=0, column=0, padx=5, pady=5)

        # Time picker (combo box for hourly increments)
        self.times = [f"{i:02}:00" for i in range(24)]
        self.time_combobox = ttk.Combobox(self.time_frame, values=self.times, state="readonly")
        self.time_combobox.grid(row=1, column=0, padx=5, pady=5)

        # Set default time
        self.time_combobox.set(self.times[self.today.hour])

    def create_calendar(self):
        """Create the calendar grid for the current month."""

        self.day_buttons = {}

        # Determine the first day of the month and the number of days in the month
        first_day = self.today.replace(day=1)
        first_day_weekday = first_day.weekday()  # Monday is 0, Sunday is 6
        num_days = self.get_days_in_month(self.today.year, self.today.month)

        # Update the title with the current month and year
        self.date_title.config(text=self.today.strftime("%B %Y"))

        # Create the calendar grid (buttons for each day of the month)
        row, col = 3, first_day_weekday  # Start at row 3 and the correct weekday column
        for day in range(1, num_days + 1):
            day_button = tk.Button(self.date_frame, text=str(day), width=4, height=2,
                                   command=lambda d=day: self.select_date(d))
            day_button.grid(row=row, column=col, padx=5, pady=5)
            self.day_buttons[day] = day_button

            # Move to the next column, and if at the end of the week, move to the next row
            col += 1
            if col > 6:
                col = 0
                row += 1

    def select_date(self, day):
        """Update the selected date."""
        self.selected_date = self.today.replace(day=day)
        print(f"Selected Date: {self.selected_date.strftime('%m/%d/%Y')}")

    def change_month(self, delta):
        """Change the month forward or backward."""
        new_month = self.today.month + delta
        if new_month < 1:
            self.today = self.today.replace(year=self.today.year - 1, month=12)
        elif new_month > 12:
            self.today = self.today.replace(year=self.today.year + 1, month=1)
        else:
            self.today = self.today.replace(month=new_month)

        # Rebuild the calendar with the updated month
        self.create_calendar()

    def change_year(self, delta):
        """Change the year forward or backward."""
        self.today = self.today.replace(year=self.today.year + delta)

        # Rebuild the calendar with the updated year
        self.create_calendar()

    def get_days_in_month(self, year, month):
        """Return the number of days in a given month."""
        # Handle December (month 12) or any month with 31 days
        if month == 12:
            return 31
        # Handle months with 30 days
        if month in [4, 6, 9, 11]:
            return 30
        # Handle February (Leap Year check)
        if month == 2:
            return 29 if (year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)) else 28
        # Handle months with 31 days
        return 31


# Create the Tkinter root window
root = tk.Tk()

# Create the DateTimePickerApp object
app = DateTimePickerApp(root)

# Run the Tkinter event loop
root.mainloop()

```
